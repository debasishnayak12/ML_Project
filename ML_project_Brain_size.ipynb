# Outliers

```python
#box plot for checking outliers
for i in df.columns:
    if df[i].dtypes!="object":
        plt.figure(figsize=(3,2))
        sns.boxplot(df[i])
        plt.xlabel(f"{i}")
        plt.show()
```

Observation : - There is no outliers in data

# Skewness

```python
#box plot for checking outliers
for i in df.columns:
    if df[i].dtypes!="object":
        plt.figure(figsize=(3,2))
        sns.histplot(df[i],kde=True)
       
        plt.show()
```

NO skewed data
